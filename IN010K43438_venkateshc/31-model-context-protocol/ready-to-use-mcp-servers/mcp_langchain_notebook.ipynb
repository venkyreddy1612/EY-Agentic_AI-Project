# 🚀 1. MCP Servers You Can Use Today

## 📁 File System MCP

### ✅ What it does
- Read/write files  
- Perfect for RAG, agents, automation  

### ▶️ Run MCP Server
```bash
pip install mcp
python -m mcp.server.filesystem --root ./data
```

### 🧠 LangChain Integration
```python
import requests
from langchain.tools import Tool
from langchain_openai import ChatOpenAI
from langchain.agents import initialize_agent

def read_file(filename):
    res = requests.post(
        "http://localhost:8000/tools/read_file",
        json={"path": filename}
    )
    return res.json()

tool = Tool(
    name="Read File",
    func=read_file,
    description="Read contents of a file"
)

llm = ChatOpenAI(model="gpt-4o-mini")

agent = initialize_agent([tool], llm, agent="zero-shot-react-description")

print(agent.run("Read notes.txt and summarize"))
```

---

## 🌐 Browser MCP (Playwright)

### ✅ What it does
- Open websites  
- Scrape content  
- Take screenshots  

### ▶️ Run MCP Server
```bash
pip install mcp-playwright
python -m mcp_playwright.server
```

### 🧠 LangChain Example
```python
def browse(url):
    res = requests.post(
        "http://localhost:8001/tools/open_page",
        json={"url": url}
    )
    return res.json()["content"]

browser_tool = Tool(
    name="Browser",
    func=browse,
    description="Fetch webpage content"
)

agent = initialize_agent([browser_tool], llm, agent="zero-shot-react-description")

print(agent.run("Summarize https://example.com"))
```

## 🗄️ PostgreSQL MCP

### ✅ What it does
- Query database using natural language  
- Great for analytics agents  

### ▶️ Run MCP Server
```bash
pip install mcp-postgres
python -m mcp_postgres.server --dsn "postgresql://user:pass@localhost/db"
```

### 🧠 LangChain Example
```python
def query_db(query):
    res = requests.post(
        "http://localhost:8002/tools/query",
        json={"query": query}
    )
    return res.json()

db_tool = Tool(
    name="Database",
    func=query_db,
    description="Query PostgreSQL database"
)

agent = initialize_agent([db_tool], llm, agent="zero-shot-react-description")

print(agent.run("What are top 5 selling products?"))
```



# 🧩 2. What These MCP Servers Actually Do

## 🧠 Core Idea

MCP servers are:

> **“A translator + executor between the AI and real-world systems”**

---

## 🔍 1. Tool Discovery

They expose tools in structured format:

```json
{
  "name": "read_file",
  "parameters": {
    "path": "string"
  }
}
```

---

## ⚙️ 2. Tool Execution

- Validate inputs  
- Call actual system (file, browser, DB)  
- Handle errors  

---

## 📤 3. Structured Response

```json
{
  "content": "File contents here..."
}
```

---

# 📁 File System MCP

### Internally:
```
LLM → "read file notes.txt"
     ↓
MCP → open("notes.txt")
     ↓
Returns content
```

---

# 🌐 Browser MCP

### Internally:
```
LLM → "open wikipedia"
     ↓
MCP → Playwright launches browser
     ↓
Extracts page content
```

---

# 🗄️ Database MCP

### Internally:
```
LLM → "top customers"
     ↓
MCP → SQL query execution
     ↓
Returns rows
```

---

# 🔥 Key Insight

### Without MCP
- You write integrations manually  

### With MCP
- Model decides:
  - which tool  
  - when to call  
  - what input
