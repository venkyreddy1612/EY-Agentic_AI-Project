## 📁 File System MCP

```

import requests
from langchain.tools import Tool
from langchain_openai import ChatOpenAI
from langchain.agents import initialize_agent

# Replace with your OpenAI key
llm = ChatOpenAI(model="gpt-4o-mini", api_key="YOUR_OPENAI_API_KEY")

def read_file(filename):
    try:
        res = requests.post(
            "http://localhost:8000/tools/read_file",
            json={"path": filename}
        )
        return res.json()
    except Exception as e:
        return str(e)

file_tool = Tool(
    name="Read File",
    func=read_file,
    description="Read contents of a file"
)

agent = initialize_agent([file_tool], llm, agent="zero-shot-react-description")

# Example run (ensure MCP server is running)
# print(agent.run("Read notes.txt and summarize"))
```

## 🌐 Browser MCP

```

def browse(url):
    try:
        res = requests.post(
            "http://localhost:8001/tools/open_page",
            json={"url": url}
        )
        return res.json().get("content", "")
    except Exception as e:
        return str(e)

browser_tool = Tool(
    name="Browser",
    func=browse,
    description="Fetch webpage content"
)

agent_browser = initialize_agent([browser_tool], llm, agent="zero-shot-react-description")

# Example run
# print(agent_browser.run("Summarize https://example.com"))
```

## 🗄️ PostgreSQL MCP

```

def query_db(query):
    try:
        res = requests.post(
            "http://localhost:8002/tools/query",
            json={"query": query}
        )
        return res.json()
    except Exception as e:
        return str(e)

db_tool = Tool(
    name="Database",
    func=query_db,
    description="Query PostgreSQL database"
)

agent_db = initialize_agent([db_tool], llm, agent="zero-shot-react-description")

# Example run
# print(agent_db.run("What are top 5 selling products?"))
```
